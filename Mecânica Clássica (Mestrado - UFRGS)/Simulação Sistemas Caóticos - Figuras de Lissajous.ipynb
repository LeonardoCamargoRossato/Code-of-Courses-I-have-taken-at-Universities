```python
import matplotlib.pyplot as plt
import numpy as np
import math
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact
import plotly.express as px

PI = math.pi
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3))

# Definindo p2 como 2 vezes Pi
p2 = 2 * np.pi

# Criando dados
x = np.linspace(0, 40, 10000)
s0a = np.sin(p2 * x * 4)
c0a = np.cos(p2 * x * 5)
s0b = np.sin(p2 * x * 3)
c0b = np.cos(p2 * x * 5)
s0c = np.sin(p2 * x * 2)
c0c = np.cos(p2 * x * 3.5)

# Plotando as curvas em cada subplot
ax1.plot(s0a, c0a, 'tab:purple')
ax1.set_title('Axis [0, 0]')
ax2.plot(s0b, c0b, 'tab:red')
ax2.set_title('Axis [0, 1]')
ax3.plot(s0c, c0c, 'tab:blue')
ax3.set_title('Axis [0, 2]')

# Configurando rótulos para cada eixo
for ax in (ax1, ax2, ax3):
    ax.set(xlabel='x-label', ylabel='y-label')

# Escondendo rótulos de eixo X e rótulos de marca de eixo Y para gráficos no topo e à direita
for ax in (ax1, ax2, ax3):
    ax.label_outer()

plt.show()
```

# Figuras Tipo1

```python
def plota_tipo1(a,b):
    fig, ax = plt.subplots()
    
    p2 = 2*PI 
    x = np.linspace(0, 40, 10000)
    s0 = np.sin(p2*x*a)
    c0 = np.cos(p2*x*b)

    ax.plot(s0, c0)
    ax.set_title('A single plot')
```

```python
@interact
def tabela_interativa( a=(0,20,1), b=(0,20,1)):
    return plota_tipo1(a,b)
```

# Figuras Tipo 2

```python
def plota_tipo2(a,b,c,d):
    fig, ax = plt.subplots()
    
    p2 = 2*PI 
    x = np.linspace(0, 40, 10000)
    s1 = np.exp(-x/a)*np.sin(p2*x*b)
    c1 = np.exp(-x/c)*np.sin(p2*x*d)

    ax.plot(s1, c1)
    ax.set_title('A single plot')
```

```python
@interact
def tabela_interativa( a=(20,40,1), b=(1,10,1),c=(10,30,1), d=(1,10,1)):
    return plota_tipo2(a,b,c,d)
```
